```python
import numpy as np
```

```python
import cv2
import os
```

```python
from PIL import Image
```

```python
dataset=[]
label=[]
```

```python
no_image_directory='/content/drive/MyDrive/no'
```

```python
yes_image_directory='/content/drive/MyDrive/yes'
```

```python
no_tumor_images=os.listdir('/content/drive/MyDrive/no')
```

```python
yes_tumor_images=os.listdir('/content/drive/MyDrive/yes')
```

```python
print(no_tumor_images)
```

```python
for i, image_name in enumerate(no_tumor_images):
  if(image_name.lower().endswith('.jpg')):
        full_path = no_image_directory + '/'+ image_name
        image = cv2.imread(full_path)
        image=Image.fromarray(image,'RGB')
        image = image.resize((64, 64))
        dataset.append(np.array(image))
        label.append(0)
```

```python
for i, image_name in enumerate(yes_tumor_images):
  if(image_name.lower().endswith('.jpg')):
    full_path= yes_image_directory + '/'+image_name
    image=cv2.imread(full_path)
    image=Image.fromarray(image,'RGB')
    image=image.resize((64,64))
    dataset.append(np.array(image))
    label.append(1)
```

```python
print(len(dataset))
print(len(label))
```
